3. Write a program to find the sum of following series:
1! +2! +3! +.... Upto n number of terms.

```python
def factorial(x):
    fact = 1
    if x == 1 or x == 0:
        return 1
    for i in range(1, x + 1):
        fact *= i
    return fact
    
n = int(input("Enter no. of terms: "))
sum_of_series = 0
for i in range(1, n + 1):
    sum_of_series += factorial(i)
print(f"Sum of the Series is {sum_of_series}")
```

4. Write a menu driven program to carry out the simple arithmetic operations like addition,
subtraction, multiplication and division between two inputted numbers.

```python
first = float(input("Enter first number:"))
second = float(input("Enter second number:"))
operation = input("""Select an operation:
                       +: Addition
                       -:Subtraction
                       *:Multiplication
                       /:Division""")
match operation:
    case "+":
        print(f"Sum of {first} and {second} is {first + second}")
    case "-":
        print(f"Subtraction of {first} and {second} is {first - second}")
    case "*":
        print(f"Multiplication of {first} and {second} is {first * second}")
    case "/":
        print(f"Division of {first} and {second} is {first / second}")
    case _:
        print("Enter a valid operation")
```

5. Write a program to find the sum of the digits of a number.

```python
num = int(input("Enter a number:"))
digits_sum = 0
temp = num
while temp > 0:
    digit = temp % 10
    digits_sum += digit
    temp //= 10
print(f"Sum of Digits of {num} is {digits_sum}")
```
