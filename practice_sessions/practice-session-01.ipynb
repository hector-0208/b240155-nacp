4. Create a program that asks for a temperature in Celsius and converts it to Fahrenheit
   using the formula:F = (C \* 9/5) + 32


```python
Celsius = int(input("Enter temperature in celsius: "))
Fahrenheit = (Celsius * 9 / 5) + 32
print(f"Fahrenheit: {Fahrenheit}")
```

5. Use the Simple Interest formula to calculate how much money someone will have in
   the bank.Formula: A = P(1 + rt)  
   P = Principal amount (initial investment)  
   r = Annual interest rate (decimal)  
   t = Time in years  
   Goal: Create variables for P, r, and t, then calculate and print the final amount A.


```python
principal = 1000
interest = 5 / 100
time = 3
Amount = principal * (1 + interest * time)
print(Amount)
```
