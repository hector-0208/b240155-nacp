```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y = np.array([-15, -25, 30, 5])

plt.plot(x, y)
plt.show()
```

```python
"""
Different markers for matplotlib:
"." - point
"," - pixel
"o" - circle
"*" - star
"$...$" - renders the string using mathtext
"P" - plus(filled)
"X" - X(filled)

Marker styles: 
markersize can be abbreviated as ms
markerfacecolor as mfc
markeredgecolor as mec

linestyles:
solid(default), dotted, dashed, dashdot, None(for no lines)
color - for line color
"""
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([-15, -25, 30, 5])
y2 = np.array([-1, 0, 20, 35])

line_style = dict(marker=".", 
        markersize=20,
        markerfacecolor="#2cfe4f",
        markeredgecolor="#b1624e",
        linestyle="solid",
        linewidth="3")

plt.plot(x, y1, **line_style, color="#e42a24")
plt.plot(x, y2, **line_style, color="#31b7d8")
plt.show()
```

```python
x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([-15, -25, 30, 5])
y2 = np.array([-1, 0, 20, 35])

plt.title("Data", fontsize=24,
          family="Times New Roman",
          fontweight="bold",
          color="#48e933")

plt.xlabel("Year",
           fontsize=16,
           family="Arial",
           fontweight="light",
           color="#2B8508")

plt.ylabel("Temperature",
           fontsize=16,
           family="Arial",
           fontweight="light",
           color="#2B8508")

plt.xticks(x)
plt.tick_params(axis="both",
                colors="#850851")

plt.plot(x, y1, color="#e42a24")
plt.plot(x, y2, color="#31b7d8")
plt.show()
```

```python
x = np.array([2023, 2024, 2025, 2026])
y = np.array([-15, -5, 30, 5])

plt.grid(axis="x",
         linewidth=2,
         color="lightgreen",
         linestyle="dotted")

plt.grid(axis="y",
         linewidth=2,
         color="lightgray",
         linestyle="dashed")

plt.plot(x, y)
plt.show()
```

```python
categories = ["Fruits", "Carbs", "Dairy", "Protein"]
values = [3, 5, 2, 6]

plt.bar(categories, values, color="skyblue")
# plt.barh(categories, values, color="skyblue") # barh() - horizontal bar

plt.title("Daily Consumption")
plt.xlabel("Food")
plt.ylabel("Quantity")

plt.show()
```

```python
categories = ["Freshmen", "Sophomores", "Juniors", "Seniors"]
values = [300, 200, 235, 130]
colors = ["red", "yellow", "blue", "green"]
""" 
autopct - percentages, %% - to display % symbol

"""
plt.pie(values, labels=categories,
        autopct="%1.1f%%",
        colors=colors,
        explode=[0, 0, 0, 0.1],
        shadow=True,
        startangle=45)

plt.title("College")
plt.show()
```
